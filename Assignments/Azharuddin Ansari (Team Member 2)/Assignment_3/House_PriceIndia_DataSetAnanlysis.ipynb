```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Uni Variate Analysis

data = pd.read_csv('/content/House Price India.csv')
print(data.head())
```


**Uni Variante Analysis**



```python
data.head()
```

```python
data.shape
```

```python
sns.histplot(data['Date'])
```

```python
sns.distplot(data.Date)
sns.displot(data.Lattitude)
```

```python

sns.countplot(data['number of bathrooms'])
```

```python
plt.pie(data.Price)
plt.show()
```

```python

x = data['number of bedrooms'].value_counts()
plt.pie(x.values,
		labels=x.index,
		autopct='%1.1f%%')
plt.show()
```

```python
sns.barplot(x=data.Date.value_counts().index,y=data.Date.value_counts())
```

```python
sns.barplot(x=data.Lattitude.value_counts().index,y=data.Lattitude.value_counts())
```

**Bi Variante Analysis**

```python
# Bi Variate Analysis

import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
sns.barplot(x=data['id'], y=data['number of bedrooms'])
plt.xticks(rotation='vertical')
```

```python

sns.scatterplot(x=data['number of floors'],
				y=data['lot area'])
```

```python
sns.lineplot(x=data.Date,y=data.Lattitude)
```

```python
sns.scatterplot(x=data.Date,y=data.Lattitude)
```

**Tri Variante Analysis**

```python
plt.boxplot(data)
```

```python
sns.jointplot(x='Built Year',y='Price',data=data)
```

```python
fig,ax = plt.subplots(2,2)
ax[0,0].plot(data.Date , data.Price)
ax[1,0].scatter(data.Date , data.Price)
ax[1,1].bar(data.Date , data.Price)
```

```python
sns.heatmap(data.corr(),annot=True)
```

```python
sns.pairplot(data)
```
